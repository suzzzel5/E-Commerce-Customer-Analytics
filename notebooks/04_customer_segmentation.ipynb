```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Visualization settings
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
# Load the cleaned data 

df = pd.read_csv("../data/cleaned_ecommerece.csv")
df.shape
```

# RFM Scoring system


```python

# Create a copy for RFM analysis
df_rfm = df.copy()
# === RECENCY SCORE ===
# Lower days = better (more recent purchase)
# We'll use quintiles (5 groups) and reverse the score
# Score 5 = most recent, Score 1 = least recent

df_rfm['R_Score'] = pd.qcut(df_rfm['days_since_last_purchase'], 
                             q=5, labels=[5, 4, 3, 2, 1])

# === FREQUENCY SCORE ===
# Higher items purchased = better
# Score 5 = most items, Score 1 = least items

df_rfm['F_Score'] = pd.qcut(df_rfm['items_purchased'].rank(method='first'), 
                             q=5, labels=[1, 2, 3, 4, 5])

# === MONETARY SCORE ===
# Higher spending = better
# Score 5 = highest spend, Score 1 = lowest spend

df_rfm['M_Score'] = pd.qcut(df_rfm['total_spend'], 
                             q=5, labels=[1, 2, 3, 4, 5])

# Convert to integers for easier calculation
df_rfm['R_Score'] = df_rfm['R_Score'].astype(int)
df_rfm['F_Score'] = df_rfm['F_Score'].astype(int)
df_rfm['M_Score'] = df_rfm['M_Score'].astype(int)

# Calculate RFM Score (simple concatenation)
df_rfm['RFM_Score'] = (df_rfm['R_Score'].astype(str) + 
                       df_rfm['F_Score'].astype(str) + 
                       df_rfm['M_Score'].astype(str))

# Calculate RFM Total (sum for easier segmentation)
df_rfm['RFM_Total'] = df_rfm['R_Score'] + df_rfm['F_Score'] + df_rfm['M_Score']

print("✅ RFM Scores calculated!")
print()

# Show examples
print("Sample RFM Scores:")
print("-" * 70)
sample = df_rfm[['customer_id', 'days_since_last_purchase', 'items_purchased', 
                 'total_spend', 'R_Score', 'F_Score', 'M_Score', 'RFM_Total']].head(10)
print(sample.to_string(index=False))
print()

# Score distribution
print("RFM Score Distribution:")
print(f"   R_Score (Recency): Min={df_rfm['R_Score'].min()}, Max={df_rfm['R_Score'].max()}, Avg={df_rfm['R_Score'].mean():.2f}")
print(f"   F_Score (Frequency): Min={df_rfm['F_Score'].min()}, Max={df_rfm['F_Score'].max()}, Avg={df_rfm['F_Score'].mean():.2f}")
print(f"   M_Score (Monetary): Min={df_rfm['M_Score'].min()}, Max={df_rfm['M_Score'].max()}, Avg={df_rfm['M_Score'].mean():.2f}")
print(f"   RFM_Total: Min={df_rfm['RFM_Total'].min()}, Max={df_rfm['RFM_Total'].max()}, Avg={df_rfm['RFM_Total'].mean():.2f}")
print()

```
